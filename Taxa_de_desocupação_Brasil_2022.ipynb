MAPA CONTENDO A LOCALIZAÇÃO DOS 5 ESTADOS COM AS MAIORES TAXA DE DESOCUPAÇÃO NO 4º TRIMESTRE DE 2022

```python
estados = ["Bahia", "Amapá", "Pernambuco", "Sergipe", "Rio de Janeiro"]
taxas_desemprego_2022 = [13.5, 13.3, 12.3, 11.9, 11.4]  # Taxas fictícias apenas para demonstração

estados_coordenadas = [
    # Bahia
    [-12.5797, -41.7007],

    # Amapá
    [1.3545, -51.9160],

    # Pernambuco
    [-8.8137, -36.9541],

    # Sergipe
    [-10.5741, -37.3857],

    # Rio de Janeiro
    [-22.9068, -43.1729]
]

# Determinar os estados com as maiores taxas de desemprego
num_maiores_taxas = 5  # Definir quantos estados considerar com as maiores taxas
maiores_indices = sorted(range(len(taxas_desemprego_2022)), key=lambda i: taxas_desemprego_2022[i], reverse=True)
maiores_estados = [estados[i] for i in maiores_indices[:num_maiores_taxas]]
maiores_taxas_desemprego = [taxas_desemprego_2022[i] for i in maiores_indices[:num_maiores_taxas]]

# Criar um mapa centralizado nas coordenadas do Brasil
mapa_estados = folium.Map(location=[-10.7799, -53.0978], zoom_start=5, control_scale=True, tiles='OpenStreetMap')

# Adicionar marcadores para cada estado no mapa, destacando os estados com maiores taxas de desemprego
for i in range(len(estados_coordenadas)):
    coordenadas_estado = estados_coordenadas[i]
    taxa_desemprego = taxas_desemprego_2022[i]

    # Definir se o estado está entre os estados com maiores taxas de desemprego
    destacar_estado = estados[i] in maiores_estados

    # Criar popup com o nome do estado e a taxa de desemprego
    popup_text = f"<b>{estados[i]}</b><br>Taxa de desemprego (4º trimestre de 2022): {taxa_desemprego:.1f}%"

    # Adicionar marcador personalizado ao mapa
    folium.Marker(
        location=coordenadas_estado,
        popup=popup_text,
        icon=folium.Icon(color='red' if destacar_estado else 'blue', icon='info-sign'),
    ).add_to(mapa_estados)

titulo = "Mapa dos Estados com as Maiores Taxas de Desemprego (4º trimestre de 2022)"
mapa_estados.get_root().html.add_child(folium.Element(f"<h1>{titulo}</h1>"))
# Exibir o mapa
mapa_estados


```
