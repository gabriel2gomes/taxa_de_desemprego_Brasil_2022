# Base de dados sobre a taxa de desocupação de pessoas de 14 anos ou mais de idade (%) - Brasil, Grande Região e Unidade da Federação

# **Análise sobre a taxa de desocupação no Brasil em 2022**

```python

df.iloc[0:1][['Brasil, Grande Região e Unidade da Federação','1º trimestre 2022','2º trimestre 2022','3º trimestre 2022','4º trimestre 2022']]
```

```python
df[['1º trimestre 2022','2º trimestre 2022', '3º trimestre 2022', '4º trimestre 2022']] = df[['1º trimestre 2022','2º trimestre 2022', '3º trimestre 2022', '4º trimestre 2022']].apply(pd.to_numeric, errors='coerce')
fig = px.bar(df[0:1],x='Brasil, Grande Região e Unidade da Federação', y=['1º trimestre 2022','2º trimestre 2022', '3º trimestre 2022', '4º trimestre 2022'], barmode='group')
fig.update_layout(title='Taxa  Desemprego do Brasil no 1º trim 2º trim 3ºtrim e 4º trim de 2022',xaxis_title='Comparação do 1º 2º,3º e 4ª trimestre de 2022',
yaxis_title='Taxa de Desemprego (%)')
fig.show()

```
