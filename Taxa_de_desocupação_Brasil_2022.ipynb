MAPA CONTENDO A LOCALIZAÇÃO DOS 5 ESTADOS COM AS MAIORES TAXA DE DESOCUPAÇÃO NO 4º TRIMESTRE DE 2022

```python


estados = ["Bahia", "Amapá", "Pernambuco", "Sergipe", "Rio de Janeiro"]
taxas_desemprego_2022 = [13.5, 13.3, 12.3, 11.9, 11.4]

estados_coordenadas = [
    # Bahia
    [-12.5797, -41.7007],

    # Amapá
    [1.3545, -51.9160],

    # Pernambuco
    [-8.0476, -34.8770],  # Corrigida a coordenada de Pernambuco

    # Sergipe
    [-10.5741, -37.3857],

    # Rio de Janeiro
    [-22.9068, -43.1729]
]

# Criar um mapa centralizado nas coordenadas do Brasil
mapa_estados = folium.Map(location=[-10.7799, -53.0978], zoom_start=5, control_scale=True, tiles='OpenStreetMap')

# Adicionando marcadores apenas para os estados no mapa, destacando os estados com maiores taxas de desemprego
for i in range(len(estados_coordenadas)):
    coordenadas_estado = estados_coordenadas[i]
    taxa_desemprego = taxas_desemprego_2022[i]

    # Definir se o estado está entre os estados com maiores taxas de desemprego
    destacar_estado = estados[i] in estados

    # Criar DivIcon com a taxa de desemprego
    if estados[i] == "Rio de Janeiro":
        div_icon = DivIcon(
            icon_size=(30, 30),
            icon_anchor=(0, -10),  # Ajuste a posição do ícone para o Rio de Janeiro
            html=f'<div style="font-size: 14px; color: {"red" if destacar_estado else "blue"};">{taxa_desemprego:.1f}%</div>'
        )
    else:
        div_icon = DivIcon(
            icon_size=(30, 30),
            icon_anchor=(15, 0),
            html=f'<div style="font-size: 14px; color: {"red" if destacar_estado else "blue"};">{taxa_desemprego:.1f}%</div>'
        )

    # Adicionar marcador personalizado ao mapa com DivIcon
    folium.Marker(
        location=coordenadas_estado,
        icon=div_icon,
        popup=folium.Popup(f"<b>{estados[i]}</b>", parse_html=True),
    ).add_to(mapa_estados)

# Exibir o mapa
mapa_estados

mapa_estados

```
